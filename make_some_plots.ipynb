```python
plt.style.use(hep.style.ROOT)
import numpy as np

def cms(ax, fontsize=20, hist2d=False):
    gap = 0.08
    if hist2d:
        gap = 0.1
    plt.text(0., 1., "CMS",
                  fontsize=fontsize,
                  horizontalalignment='left',
                  verticalalignment='bottom',
                  transform=ax.transAxes,
                  weight="bold"
                 )
    plt.text(gap, 1., "Simulation Preliminary",
                  fontsize=fontsize,
                  horizontalalignment='left',
                  verticalalignment='bottom',
                  transform=ax.transAxes,
                  style='italic',
                 )
    return

dense=False
scale="log"
```
